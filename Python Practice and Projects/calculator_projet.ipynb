```python
import math

# Function to handle arithmetic operations
def calculate(a, operator, b):
    try:
        if operator == "+":
            return f"{a} + {b} = {a + b}"
        elif operator == "-":
            return f"{a} - {b} = {a - b}"
        elif operator == "x":
            return f"{a} x {b} = {a * b}"
        elif operator == "/":
            return f"{a} / {b} = {a / b}"
        elif operator == "%":
            return f"{a} % {b} = {a % b}"
        elif operator == "^":
            return f"{a} ^ {b} = {a ** b}"
        elif operator == "sqrt":
            return f"sqrt({a}) = {math.sqrt(a)}"
        else:
            return "Invalid operator. Please use one of +, -, x, /, %, ^, sqrt."
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."

# Function to handle file operations
def handle_file_operations():
    while True:
        file_name = input("Enter the file name: ")
        try:
            with open(file_name, 'r') as file:
                print("\nFile Content:\n")
                for line in file:
                    print(line.strip())
            break
        except FileNotFoundError:
            print("Error: File not found. Please try again.")

# Main calculator loop
def main():
    history = []  # List to store calculation history
    while True:
        try:
            user_input1 = input("Enter your first number or 'stop' to stop: ").lower()
            if user_input1 == "stop":
                next_action = input("Enter 'file' to view a file or 'back' to return: ").lower()
                if next_action == "file":
                    handle_file_operations()
                elif next_action == "back":
                    continue
                else:
                    print("Invalid input. Returning to main menu.")
                    continue
            elif user_input1 == "sqrt":
                a = float(input("Enter the number to find the square root: "))
                result = calculate(a, "sqrt", None)
                print(result)
                history.append(result)
                continue

            # Convert first number to integer
            a = float(user_input1)
            operator = input("Enter operator (+, -, x, /, %, ^,'sqrt' for square root): ")
            b = float(input("Enter your second number: "))

            # Perform calculation
            result = calculate(a, operator, b)
            print(result)
            history.append(result)

            # Write result to file
            with open("Table.txt", "a") as file:
                file.write(result + '\n')

        except ValueError:
            print("Error: Invalid input. Please enter a valid number.")
            with open("error_log.txt", "a") as error_file:
                error_file.write("ValueError: Invalid number input.\n")
        except Exception as e:
            print(f"Unexpected error: {e}")
            with open("error_log.txt", "a") as error_file:
                error_file.write(f"Unexpected error: {e}\n")

# Run the program
if __name__ == "__main__":
    main()
```
